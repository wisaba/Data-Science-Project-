```python
import pandas as pd
titanic3 = pd.read_csv('titanic.csv')
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
titanic3['Sex'] = le.fit_transform(titanic3['Sex'])
titanic3.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
titanic3 = pd.read_csv('titanic.csv')

ohe = OneHotEncoder()
categorical_columns = ["Sex"]
categorical_data = titanic3[categorical_columns]
encoded_data = ohe.fit_transform(categorical_data)
titanic4 = pd.DataFrame(encoded_data.toarray(),columns=ohe.get_feature_names_out())
pd.concat([titanic3,titanic4],axis=1)
```

```python
import pandas as pd 
titanic5 = pd.get_dummies(titanic3, columns=['Sex'],dtype=int)
titanic5
```
