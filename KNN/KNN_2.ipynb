```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
titanic_data = pd.read_csv("titanic.csv")
titanic_data.isnull().sum()
```

```python
titanic_data.drop(['Cabin'], axis = 1 , inplace=True)
titanic_data
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data['Age'].fillna(titanic_data['Age'].mean(),inplace=True)
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0],inplace=True)
```

```python
titanic_data
```

```python
le = LabelEncoder()
titanic_data["Sex"] = le.fit_transform(titanic_data["Sex"])
```

```python
titanic_data["Embarked"] = le.fit_transform(titanic_data["Embarked"])
```

```python
titanic_data
```

### train a KNN model 

```python
features = ["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]
X = titanic_data[features]
y = titanic_data["Survived"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.18)
```

```python
k = 5
my_k_near = KNeighborsClassifier(n_neighbors=k)
```

```python
my_k_near.fit(X_train,y_train)
```

```python
y_pred = my_k_near.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
```
