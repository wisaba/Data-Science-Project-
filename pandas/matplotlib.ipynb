```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
t = np.linspace(0,np.pi,200)
x = np.sin(t)
y = np.cos(t)
z = np.sin(t)**2 + np.cos(t)**2
axe1 = plt.subplot(1,3,1)
axe1.plot(t,x)
axe2 = plt.subplot(1,3,2)
axe2.plot(t,y)
axe3 = plt.subplot(1,3,3)
axe3.plot(t,z)
plt.xlabel('t')
plt.ylabel('x')
plt.show()
```

```python
t = np.arange(0,2*np.pi,0.01)
plt.rcParams['font.family']="Times new Roman"
plt.rcParams['font.size']=14
r1=1
r2=2
x = r1*np.sin(t)
y = r1*np.cos(t)
x2 = r2*np.sin(t)
y2 = r2*np.cos(t)
plt.plot(x,y,"--b",linewidth=2.5)
plt.plot(x2,y2,"--r",linewidth=2.5)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim()
plt.grid(-1)
plt.axis('equal   ')
plt.legend(["First circle","Second circle"],loc="lower right")
plt.savefig('circles.png',format='png',dpi=200)
plt.show()
```
